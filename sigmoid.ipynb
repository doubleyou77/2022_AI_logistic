### Sigmoid 구현

$$S(x) = \frac{1}{1+e^{-x}}

```python
import numpy as np
class Sigmoid:
    def __init__(self):
        self.out = None

    def forward(self, x):
        out = 1/(1+np.exp(-x))
        self.out = out
        return out
```

```python
import matplotlib.pylab as plt

s1 = Sigmoid()
x = np.arange(-10,10,0.01)
y = s1.forward(x)
plt.plot(x,y)
plt.show()
```

### 미분 어느 한 시점의 변화량
#### 머신러닝에서 미분을 사용하는 이유 특정 지점에서의 기울기를 알 수 있기 때문에

### Sigmoid 미분 추가

```python
import numpy as np
class Sigmoid:
    def __init__(self):
        self.out = None

    def forward(self, x):
        out = 1/(1+np.exp(-x))
        self.out = out
        return out

    def backward(self, dout):
        dx = dout*self.out*(1.0-self.out);
        return dx
```

```python
import matplotlib.pylab as plt

s1 = Sigmoid()
x = np.arange(-10,10,0.01)
out = np.ones(len(x))
y = s1.forward(x)
dx = s1.backward(out)
plt.plot(x,y)
plt.plot(x,dx)
plt.show()
```
